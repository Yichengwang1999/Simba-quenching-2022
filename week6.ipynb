```python
def progen_index(obj,compare,snap_path):
    
    progen_index = caesar.progen.progen_finder(obj,compare, '/disk04/rad/sim/m100n1024/s50/Groups/m100n1024_078.hdf5',snap_path ,n_most=1, save=False)
    progen_index = progen_index.flatten()
    
    return progen_index
```

```python
def loop_files_index(loop_id,input_current,snap_dir):
    index_all_snapshot = []
    obj_current = caesar.load(input_current)
    galaxies = np.array(progen_index(obj_current,obj_current,snap_dir))
    

    quenched_galaxies = []
    for g in range(0,len(galaxies)):
        if np.log10(obj_current.galaxies[g].masses['stellar'])>10.5:
            if (np.log10(float(obj_current.galaxies[g].sfr) + 1.e-3)- np.log10(obj_current.galaxies[g].masses['stellar'])) < -11:
                quenched_galaxies.append(g)


    for infile in loop_id:
        sim = caesar.load('/disk04/rad/sim/m100n1024/s50/Groups/m100n1024_%.03d.hdf5'%(infile))
        index_01 = progen_index(obj_current,sim,snap_dir)

        selected_01 = []
        for id in quenched_galaxies:
            selected_01.append(index_01[id])

        
        index_all_snapshot.append(selected_01)

    index_all_snapshot = np.array(index_all_snapshot)

    return index_all_snapshot




            

```

```python
def history_galaxy(loop_id,input_current,snap_dir,index_array,parttype):

    obj_current = caesar.load(input_current)
    galaxies = np.array(progen_index(obj_current,obj_current,snap_dir))
    
    quenched_galaxies = []
    for g in range(0,len(galaxies)):
        if np.log10(obj_current.galaxies[g].masses['stellar'])>10.5:
            if (np.log10(float(obj_current.galaxies[g].sfr) + 1.e-3)- np.log10(obj_current.galaxies[g].masses['stellar'])) < -11:
                quenched_galaxies.append(g)
    

    history_galaxies = []
    snapshot = 0
    for infile in loop_id:
        sim = caesar.load('/disk04/rad/sim/m100n1024/s50/Groups/m100n1024_%.03d.hdf5'%(infile))
        
        

        selected = np.array(index_array[snapshot])
        selected = selected[selected[:,0] != -1]

        pmass_quenched = np.log10(np.asarray([sim.galaxies[i].masses['stellar'] for i in selected[:,0]]))
        
        if parttype == 'mHI' : 
            property_quenched = np.log10(np.asarray([sim.galaxies[i].masses['HI'] for i in selected[:,0]])+1.e-3) - pmass_quenched
            
        elif parttype == 'mbh' : 
            property_quenched = np.log10(np.asarray([sim.galaxies[i].masses['bh'] for i in selected[:,0]])+1.e-3) - pmass_quenched
            
        elif parttype == 'sfr' : 
            property_quenched = np.log10(np.asarray([sim.galaxies[i].sfr for i in selected[:,0]])+1.e-3) - pmass_quenched
            
        elif parttype == 'mH2' : 
            property_quenched = np.log10(np.asarray([sim.galaxies[i].masses['H2'] for i in selected[:,0]])+1.e-3) - pmass_quenched
            
            
        elif parttype == 'mbhfr':
            property_quenched = np.log10(np.asarray([sim.galaxies[i].bhmdot for i in selected[:,0]])+1.e-3) - np.log10(np.asarray([sim.galaxies[i].sfr for i in selected[:,0]])+1.e-3)

        id_loop = [infile] * len(selected[:,0])

        history_galaxies.append([property_quenched,pmass_quenched,selected[:,1],id_loop])

        snapshot += 1
   
   
    return history_galaxies



    
```
